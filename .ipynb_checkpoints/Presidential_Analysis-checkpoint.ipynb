```python
def remark_word_count_analysis(remarks, json_file_path, csv_file_path):
    no_words = 0
    word_count_dictionary = {}
    word_count_df = pd.DataFrame(columns=['Word', 'Count'])
    for i in range(len(remarks)):
        remark = remarks[i]

        if 'text' in remark:
            remark_text = remark['text']

            # Filtering
            remark_text = remark_text.replace('\n', ' ')
            remark_text = remark_text.replace('.', ' .')
            remark_text = remark_text.replace(';', ' ;')
            remark_text = remark_text.replace(':', ' :')
            remark_text = remark_text.replace(',', ' ,')
            remark_text = remark_text.replace('?', ' ?')
            remark_text = remark_text.replace('(', '')
            remark_text = remark_text.replace(')', '')
            remark_text = remark_text.replace('’', '\'')
            remark_text = remark_text.replace('“', '')
            remark_text = remark_text.replace(u'\xa0', u' ')
            remark_words = remark_text.split(' ')

            for word in remark_words:
                no_words += 1
                word = word.lower()
                if word not in stop_words and word not in filter_words and len(word) > 3:
                    if word in word_count_dictionary:
                        word_count_dictionary[word] += 1

                        df_index = word_count_df[word_count_df['Word'] == word].index.values.astype(int)[0]
                        word_count_df.loc[df_index, 'Count'] = word_count_df.loc[df_index, 'Count'] + 1

                    else:
                        word_count_dictionary[word] = 1
                        word_count_df = word_count_df.append({"Word": word, "Count": 1}, ignore_index=True)
    word_count_dictionary['#_total_word_count'] = no_words

    # Save word count dictionary to json file
    with open(json_file_path, 'w') as file:
        json.dump(word_count_dictionary, file, indent=4, ensure_ascii=True)

    word_count_df = word_count_df.sort_values(by=['Count'], ascending=False)
    word_count_df.to_csv(csv_file_path)
```

```python
def tweet_word_count_analysis(tweets, json_file_path, csv_file_path):
    no_words = 0
    word_count_dictionary = {}
    word_count_df = pd.DataFrame(columns=['Word', 'Count'])
    for i in range(len(tweets)):
        tweet = tweets[i]
        if 'text' in tweet:
            tweet_text = tweet['text']

            # Filtering
            tweet_text = tweet_text.replace('\n', ' ')
            tweet_text = tweet_text.replace('.', ' .')
            tweet_text = tweet_text.replace(';', ' ;')
            tweet_text = tweet_text.replace(':', ' :')
            tweet_text = tweet_text.replace(',', ' ,')
            tweet_text = tweet_text.replace('?', ' ?')
            tweet_text = tweet_text.replace('(', '')
            tweet_text = tweet_text.replace(')', '')
            tweet_text = tweet_text.replace('’', '\'')
            tweet_text = tweet_text.replace('“', '')
            tweet_text = tweet_text.replace(u'\xa0', u' ')
            
            tweet_words = tweet_text.split(' ')

            for word in tweet_words:
                no_words += 1
                word = word.lower()
                if word not in stop_words and word not in filter_words and len(word) > 3:
                    if word in word_count_dictionary:
                        word_count_dictionary[word] += 1

                        df_index = word_count_df[word_count_df['Word'] == word].index.values.astype(int)[0]
                        word_count_df.loc[df_index, 'Count'] = word_count_df.loc[df_index, 'Count'] + 1

                    else:
                        word_count_dictionary[word] = 1
                        word_count_df = word_count_df.append({"Word": word, "Count": 1}, ignore_index=True)
    word_count_dictionary['#_total_word_count'] = no_words

    # Save word count dictionary to json file
    with open(json_file_path, 'w') as file:
        json.dump(word_count_dictionary, file, indent=4, ensure_ascii=True)

    word_count_df = word_count_df.sort_values(by=['Count'], ascending=False)
    word_count_df.to_csv(csv_file_path)
```

```python
def visualize_word_count_analysis(json_file_path, csv_file_path, president_name):
    json_remarks_file = open(json_file_path)
    json_remarks = json.load(json_remarks_file)
    print(f"Total number of words: {json_remarks['#_total_word_count']}")
    
    df = pd.read_csv(csv_file_path)     
    fig = plt.figure(figsize=(60,10))
    plt.bar(df['Word'][0:30], df['Count'][0:30])
    plt.xlabel('Words', fontsize=18)
    plt.ylabel('Count', fontsize=18)
    plt.xticks(fontsize=23)
    plt.title(f'Most Common Words - {president_name}', fontsize=20)
    plt.show()
```

### Grammar Analyses
https://spacy.io/api/annotation

```python
def track_word_in_dataframe_and_dict(word, track_dict, track_df):
    if word in track_dict:
        track_dict[word] += 1
        df_index = track_df[track_df['Word'] == word].index.values.astype(int)[0]
        track_df.loc[df_index, 'Count'] = track_df.loc[df_index, 'Count'] + 1
    else:
        track_dict[word] = 1
        track_df = track_df.append({"Word": word, "Count": 1}, ignore_index=True)
        
    return track_dict, track_df

def remark_grammar_analysis(remarks, pres_name):
    
    noun_count_df = pd.DataFrame(columns=['Word', 'Count'])
    noun_count_dict = {}

    adv_count_df = pd.DataFrame(columns=['Word', 'Count'])
    adv_count_dict = {}

    verb_count_df = pd.DataFrame(columns=['Word', 'Count'])
    verb_count_dict = {}
    
    adj_count_df = pd.DataFrame(columns=['Word', 'Count'])
    adj_count_dict = {}
    
    pnoun_count_df = pd.DataFrame(columns=['Word', 'Count'])
    pnoun_count_dict = {}
    
    ent_org_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_org_count_dict = {}
    
    ent_gpe_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_gpe_count_dict = {}
    
    ent_money_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_money_count_dict = {}
    
    ent_person_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_person_count_dict = {}
    
    ent_norp_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_norp_count_dict = {}
    
    ent_event_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_event_count_dict = {}
    
    ent_prod_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_prod_count_dict = {}
    
    ent_law_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_law_count_dict = {}
    
    ent_fac_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_fac_count_dict = {}
    
    ent_art_count_df = pd.DataFrame(columns=['Word', 'Count'])
    ent_art_count_dict = {}
    
    
    for i in range(len(remarks)):
        remark = remarks[i]

        if 'text' in remark:
            remark_text = remark['text']

            # Clean up text
            remark_text = remark_text.replace('\n', ' ')
            remark_text = remark_text.replace('.', ' .')
            remark_text = remark_text.replace(';', ' ;')
            remark_text = remark_text.replace(':', ' :')
            remark_text = remark_text.replace(',', ' ,')
            remark_text = remark_text.replace('?', ' ?')
            remark_text = remark_text.replace('(', '')
            remark_text = remark_text.replace(')', '')
            remark_text = remark_text.replace('’', '\'')
            remark_text = remark_text.replace('“', '')
            remark_text = remark_text.replace(u'\xa0', u' ')
            
            # Extract entities via Spacey
            nlp_doc = nlp(remark_text)
            for token in nlp_doc:
                
                word = token.text.lower()
                if len(word) <= 1:
                    continue
                                
                if token.pos_ == 'VERB':
                    (verb_count_dict,verb_count_df) = track_word_in_dataframe_and_dict(word, verb_count_dict, verb_count_df)
                    
                elif token.pos_ == 'ADJ':
                    (adj_count_dict, adj_count_df) = track_word_in_dataframe_and_dict(word, adj_count_dict, adj_count_df)

                elif token.pos_ == 'NOUN':
                    (noun_count_dict, noun_count_df) = track_word_in_dataframe_and_dict(word, noun_count_dict, noun_count_df)

                elif token.pos_ == 'ADV':
                    (adv_count_dict, adv_count_df) = track_word_in_dataframe_and_dict(word, adv_count_dict, adv_count_df)

                elif token.pos_ == 'PROPN':
                    (pnoun_count_dict, pnoun_count_df) = track_word_in_dataframe_and_dict(word, pnoun_count_dict, pnoun_count_df)

            for ent in nlp_doc.ents:
                word = ent.text.lower()
                if len(word) <= 1:
                    continue
                    
                if ent.label_ == 'ORG':
                    (ent_org_count_dict, ent_org_count_df) = track_word_in_dataframe_and_dict(word, ent_org_count_dict, ent_org_count_df)

                elif ent.label_ == 'GPE':
                    (ent_gpe_count_dict, ent_gpe_count_df) = track_word_in_dataframe_and_dict(word, ent_gpe_count_dict, ent_gpe_count_df)

                elif ent.label_ == 'MONEY':
                    (ent_money_count_dict, ent_money_count_df) = track_word_in_dataframe_and_dict(word, ent_money_count_dict, ent_money_count_df)
                    
                elif ent.label_ == 'PERSON':
                    (ent_person_count_dict, ent_person_count_df) = track_word_in_dataframe_and_dict(word, ent_person_count_dict, ent_person_count_df)
                    
                elif ent.label_ == 'NORP':
                    (ent_norp_count_dict, ent_norp_count_df) = track_word_in_dataframe_and_dict(word, ent_norp_count_dict, ent_norp_count_df)
                    
                elif ent.label_ == 'PRODUCT':
                    (ent_prod_count_dict, ent_prod_count_df) = track_word_in_dataframe_and_dict(word, ent_prod_count_dict, ent_prod_count_df)
                    
                elif ent.label_ == 'LAW':
                    (ent_law_count_dict, ent_law_count_df) = track_word_in_dataframe_and_dict(word, ent_law_count_dict, ent_law_count_df)
                    
                elif ent.label_ == 'FAC':
                    (ent_fac_count_dict, ent_fac_count_df) = track_word_in_dataframe_and_dict(word, ent_fac_count_dict, ent_fac_count_df)
                    
                elif ent.label_ == 'ART':
                    (ent_art_count_dict, ent_art_count_df) = track_word_in_dataframe_and_dict(word, ent_art_count_dict, ent_art_count_df)
                    
    # Save grammar dictionaries to json file
    with open(f'{pres_name}_grammar/grammar.json', 'w') as file:
        grammar_dict = {
            "verb":verb_count_dict,
            "adv":adv_count_dict,
            "noun":noun_count_dict,
            "p_noun":pnoun_count_dict,
            "adj":adj_count_dict,
            "ent_org":ent_org_count_dict,
            "ent_gpe":ent_gpe_count_dict,
            "ent_money":ent_money_count_dict,
            "ent_person":ent_person_count_dict,
            "ent_norp":ent_norp_count_dict,
            "ent_product":ent_prod_count_dict,
            "ent_law":ent_law_count_dict,
            "ent_fac":ent_fac_count_dict,
            "ent_art":ent_art_count_dict

        }
        json.dump(grammar_dict, file, indent=4, ensure_ascii=True)
    
    # Sort dataframe by word count and save
    df_array = [verb_count_df, noun_count_df, adv_count_df, adj_count_df, pnoun_count_df, ent_org_count_df, ent_gpe_count_df, ent_money_count_df, ent_person_count_df, ent_norp_count_df, ent_prod_count_df, ent_law_count_df, ent_fac_count_df, ent_art_count_df]
    save_name_array = ['verb', 'noun', 'adv', 'adj', 'pnoun', 'ent_org', 'ent_gpe', 'ent_money', "ent_person", "ent_norp", "ent_product", "ent_law", "ent_fac", "ent_art"]
    
    for i in range(len(save_name_array)):
        df_grammer = df_array[i]
        df_grammer = df_grammer.sort_values(by=['Count'], ascending=False)
        df_grammer.to_csv(f'{pres_name}_grammar/{save_name_array[i]}.csv')  
```

```python
def visualize_grammar_analysis(president_name):
    
    verb_df = pd.read_csv(f'{president_name}_grammar/verb.csv') 
    noun_df = pd.read_csv(f'{president_name}_grammar/noun.csv') 
    adv_df = pd.read_csv(f'{president_name}_grammar/adv.csv') 
    adj_df = pd.read_csv(f'{president_name}_grammar/adj.csv') 
    pnoun_df = pd.read_csv(f'{president_name}_grammar/pnoun.csv') 
    ent_org_df = pd.read_csv(f'{president_name}_grammar/ent_org.csv') 
    ent_gpe_df = pd.read_csv(f'{president_name}_grammar/ent_gpe.csv') 
    ent_money_df = pd.read_csv(f'{president_name}_grammar/ent_money.csv') 
    ent_person_df = pd.read_csv(f'{president_name}_grammar/ent_person.csv') 
    ent_norp_df = pd.read_csv(f'{president_name}_grammar/ent_norp.csv') 
    ent_product_df = pd.read_csv(f'{president_name}_grammar/ent_product.csv') 
    ent_law_df = pd.read_csv(f'{president_name}_grammar/ent_law.csv') 
    ent_fac_df = pd.read_csv(f'{president_name}_grammar/ent_fac.csv') 
    ent_art_df = pd.read_csv(f'{president_name}_grammar/ent_art.csv') 

    grammar_array = ['verb', 'noun', 'adv', 'adj', 'pnoun', 'ent_org', 'ent_gpe', 'ent_money', 'ent_person', 'ent_norp', 'ent_product', 'ent_law', 'ent_fac', 'ent_art']
    grammar_df_array = [verb_df, noun_df, adv_df, adj_df, pnoun_df, ent_org_df, ent_gpe_df, ent_money_df, ent_person_df, ent_norp_df, ent_product_df, ent_law_df, ent_fac_df, ent_art_df]

    for i in range(len(grammar_array)): 
        df = grammar_df_array[i]
        gram = grammar_array[i]
        
        fig = plt.figure(figsize=(40,5))
        plt.bar(df['Word'][0:24], df['Count'][0:24])
        plt.xlabel(f'{gram.capitalize()}', fontsize=18)
        plt.xticks(fontsize=15)
        plt.ylabel('Count', fontsize=18)
        plt.title(f'{gram.capitalize()} - {president_name.capitalize()}', fontsize=25)
        plt.show()
```

### Impeachment Specific Analyses

```python
impeachment_words = ['impeachment', 'impeach', 'supermajority', 'inquiry', 'indictment']
def impeachment_related_analysis(remarks, president_name):
    
    word_count_dictionary = {'impeachment':0, 'impeach': 0, 'supermajority':0, 'inquiry':0, 'indictment':0}
    word_count_df = pd.DataFrame(columns=['Word', 'Count'])
    word_count_df = word_count_df.append({'Word':'impeachment', 'Count':0}, ignore_index=True)
    word_count_df = word_count_df.append({'Word':'impeach', 'Count':0}, ignore_index=True)
    word_count_df = word_count_df.append({'Word':'supermajority', 'Count':0}, ignore_index=True)
    word_count_df = word_count_df.append({'Word':'inquiry', 'Count':0}, ignore_index=True)
    word_count_df = word_count_df.append({'Word':'indictment', 'Count':0}, ignore_index=True)
    
    for i in range(len(remarks)):
        remark = remarks[i]

        if 'text' in remark:
            remark_text = remark['text']

            # Filtering
            remark_text = remark_text.replace('\n', ' ')
            remark_text = remark_text.replace('.', ' .')
            remark_text = remark_text.replace(';', ' ;')
            remark_text = remark_text.replace(':', ' :')
            remark_text = remark_text.replace(',', ' ,')
            remark_text = remark_text.replace('?', ' ?')
            remark_text = remark_text.replace('(', '')
            remark_text = remark_text.replace(')', '')
            remark_text = remark_text.replace('’', '\'')
            remark_text = remark_text.replace('“', '')
            remark_text = remark_text.replace(u'\xa0', u' ')
            remark_words = remark_text.split(' ')

            for word in remark_words:
                word = word.lower()
                if word in impeachment_words:
                    if word in word_count_dictionary:
                        word_count_dictionary[word] += 1

                        df_index = word_count_df[word_count_df['Word'] == word].index.values.astype(int)[0]
                        word_count_df.loc[df_index, 'Count'] = word_count_df.loc[df_index, 'Count'] + 1


    # Save word count dictionary to json file
    with open(f'{president_name}_impeachment/impeachment_words.json', 'w') as file:
        json.dump(word_count_dictionary, file, indent=4, ensure_ascii=True)

    word_count_df.to_csv(f'{president_name}_impeachment/impeachment_words.csv')
```

```python
def visualize_impeachment_analysis(president_name):
    df = pd.read_csv(f'{president_name}_impeachment/impeachment_words.csv')     
    fig = plt.figure(figsize=(30,10))
    plt.bar(df['Word'][0:5], df['Count'][0:5])
    plt.xlabel('Words', fontsize=18)
    plt.ylabel('Count', fontsize=18)
    plt.rc('xtick', labelsize=30) 
    plt.title(f'Impeachment Words - {president_name}', fontsize=25)
    plt.show()
```

### Toxicity Analyses

```python
def visualize_toxicity_analysis(president_name):
    
    json_remarks_file = open(f'{president_name}_remarks_w_toxicity.json')
    json_remarks = json.load(json_remarks_file)
    toxicity_score_array = []

    for i in range(len(json_remarks)): 
        remark = json_remarks[i]
        if 'toxicity_score' in remark:
            toxicity_score_array.append(remark['toxicity_score'])
        else:
            toxicity_score_array.append(0.0)

    print(f'{president_name.capitalize()} Mean Toxicity: {np.mean(toxicity_score_array, axis=0)}')
          
    fig = plt.figure(figsize=(20,10))
    plt.plot(toxicity_score_array)
    plt.xlabel('Remarks', fontsize=18)
    plt.ylabel('Toxicity Score (0-1)', fontsize=18)
    plt.rc('xtick', labelsize=30) 
    plt.title(f'Toxicity Score - {president_name.capitalize()}', fontsize=25)
    plt.show()
```

```python
import scattertext as st
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.snowball import SnowballStemmer
import re
stemmer = SnowballStemmer("english")
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from spacy.lemmatizer import Lemmatizer
from spacy.lookups import Lookups
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
def filter_remark(remark_text):
    filtered_remark = ""
    remark_text = remark_text.replace('\n', ' ')
    remark_text = remark_text.replace('.', ' .')
    remark_text = remark_text.replace(';', ' ;')
    remark_text = remark_text.replace(':', ' :')
    remark_text = remark_text.replace(',', ' ,')
    remark_text = remark_text.replace('?', ' ?')
    remark_text = remark_text.replace('(', '')
    remark_text = remark_text.replace(')', '')
    remark_text = remark_text.replace('’', '\'')
    remark_text = remark_text.replace('“', '')
    remark_text = remark_text.replace(u'\xa0', u' ')
    remark_words = remark_text.split(' ')

    for word in remark_words:
        word = word.lower()
        if word not in spacy_stopwords and len(word) > 3:
            filtered_remark += word + " "
            
    filtered_doc = nlp(filtered_remark)
    lemma_remark = " ".join([token.lemma_ for token in filtered_doc])
    return lemma_remark
```

### PCA Analyses

```python
def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
```

```python
trump_remarks_array = []
for i in range(len(trump_remarks)):
    remark = trump_remarks[i]
    text = trump_remarks[i]['text']
    text = filter_remark(text)
    trump_remarks_array.append(text)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 min_df=0.2, stop_words='english',
                                 use_idf=True, tokenizer=tokenize_and_stem, ngram_range=(1,3),smooth_idf=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(trump_remarks_array) #fit the vectorizer to synopses
print(tfidf_matrix.shape)
```

```python
terms = tfidf_vectorizer.get_feature_names()
print(terms)
```

```python
true_k = 3
kmeans = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
kmeans.fit(tfidf_matrix)
```

```python
print("Top terms per cluster:")
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = tfidf_vectorizer.get_feature_names()
for i in range(true_k):
    print(f'Cluster: {i}')
    for ind in order_centroids[i, :10]:
        print (terms[ind])
    print("-----------------")
```

```python
pca = PCA(n_components=2).fit(tfidf_matrix.todense())
data2D = pca.transform(tfidf_matrix.todense())
plt.scatter(data2D[:,0], data2D[:,1])
plt.show()
```

```python
centers2D = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers2D[:,0], centers2D[:,1], 
            marker='x', s=200, linewidths=3, c='r')
plt.scatter(data2D[:,0], data2D[:,1])
plt.show()  
```

```python
def find_children_adjectives(token):
    adjectives = []
    if token.pos_ == 'ADJ':
        if token.text not in adjectives:
            adjectives.append(token.text)
    
    for child in token.children:
        if child.pos_ == 'ADJ':
            if token.text not in adjectives:
                adjectives.append(child.text)
            adjectives += find_children_adjectives(child)
        else:
            adjectives += find_children_adjectives(child)
    return adjectives
```

```python
def find_adjectives(token):
    adjectives = []
    for ancestor in token.ancestors:
        if ancestor.pos_ == 'ADJ':
            if ancestor.text not in adjectives:
                adjectives.append(ancestor.text)
            adjectives += find_children_adjectives(ancestor)
        else:
            adjectives += find_children_adjectives(ancestor)
    return adjectives
```
